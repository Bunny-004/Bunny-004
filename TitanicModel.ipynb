```python
import pandas as pd
```

# Data Exploring

```python
t_data= pd.read_csv('train.csv')
t_data.head()
# Reading the Dataset.
```

# Data Modification

```python
ports= pd.get_dummies(t_data.Embarked, prefix= 'Port')
ports.head()
# Creating 3 attribute to represent the attribute- 'Embarked'
```

```python
t_data= t_data.join(ports)
t_data.head()
# Adding the 3 attribute that we just made.
```

```python
t_data= t_data.drop('Embarked', axis=1)
t_data.head()
# Deleting the 'Embarked' attribute from the data.
```

```python
t_data.Sex= t_data.Sex.map({'male':0,'female':1})
t_data.head()
```

```python
sur_data= t_data.Survived.copy()
mod_data= t_data.drop(['Survived'], axis=1)
mod_data.head()
# Cutting (Storing and Removing) the 'Survived' attribute for training purposes.
```

```python
mod_data.drop(['PassengerId','Name','Ticket','Cabin'], axis= 1, inplace= True)
mod_data.head()
# Removing arbitary attributes from the data.
```

```python
mod_data.isnull().values.any()
# Boolean Checking for null values in the data.
```

```python
mod_data[pd.isnull(mod_data).any(axis= 1)]
# Finding where the values are null (NaN).
```

```python
mod_data.fillna(mod_data.Age.mean(), inplace= True)
mod_data.head()
# Filling the null (NaN) values with mean values.
```

```python
mod_data.isnull().values.any()
# Re-checking for null values.
```

# **Model Building**

```python
from sklearn.model_selection import train_test_split
mod_data_train, mod_data_valid, sur_data_train, sur_data_valid = train_test_split(mod_data,sur_data,test_size= 0.2, random_state= 7)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
```

```python
model.fit(mod_data_train, sur_data_train)
```

# Model Checking

```python
model.score(mod_data_train, sur_data_train)
```

```python
model.score(mod_data_valid, sur_data_valid)
```
